```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
policy = np.random.randint(0, 5, size=(10, 10))
productions = np.random.randint(0, 10, size=(10, 10))
strengths = np.random.randint(0, 10, size=(10, 10))
```

```python
fig = plt.figure(dpi=120)
ax = fig.gca()

ax.matshow(productions, cmap="magma")
ax.scatter(*np.where(policy==0), marker="^", color="red", s=strengths[policy==0])
ax.scatter(*np.where(policy==1), marker=">", color="red")
ax.scatter(*np.where(policy==2), marker="v", color="red")
ax.scatter(*np.where(policy==3), marker="<", color="red")
```
